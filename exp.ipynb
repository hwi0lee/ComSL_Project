```python
import os
import shutil
import random

# 폴더 경로 설정
source_wav_folder = "./data/Training/01.원천데이터"
source_json_folder = "./data/Training/02.라벨링데이터"
destination_wav_folder = "./data/Validation/01.원천데이터"
destination_json_folder = "./data/Validation/02.라벨링데이터"

# 대상 폴더가 없으면 생성
os.makedirs(destination_wav_folder, exist_ok=True)
os.makedirs(destination_json_folder, exist_ok=True)

# 모든 .wav 파일 리스트 가져오기
wav_files = [f for f in os.listdir(source_wav_folder) if f.endswith(".wav")]

# 전체 파일 중 10%를 랜덤하게 선택
num_files_to_move = max(1, int(len(wav_files) * 0.1))  # 최소 1개는 이동
selected_files = random.sample(wav_files, num_files_to_move)

# 파일 이동
for wav_file in selected_files:
    # WAV 파일 경로 설정
    source_wav_path = os.path.join(source_wav_folder, wav_file)
    dest_wav_path = os.path.join(destination_wav_folder, wav_file)
    
    # 대응하는 JSON 파일 이름 설정
    json_file = wav_file.replace(".wav", ".json")
    source_json_path = os.path.join(source_json_folder, json_file)
    dest_json_path = os.path.join(destination_json_folder, json_file)

    # WAV 파일 이동
    if os.path.exists(source_wav_path):
        shutil.move(source_wav_path, dest_wav_path)
        print(f"Moved WAV: {wav_file}")
    else:
        print(f"Missing WAV file: {wav_file}")

    # JSON 파일 이동
    if os.path.exists(source_json_path):
        shutil.move(source_json_path, dest_json_path)
        print(f"Moved JSON: {json_file}")
    else:
        print(f"Missing JSON file: {json_file}")

print(f"Moved {len(selected_files)} files to Validation folders.")
```

```python
import os
import torchaudio

# 파일이 있는 디렉토리 경로
directory = "./dataset/train/clipped_mp3"

# 오류가 발생하는 파일을 저장할 리스트
error_files = []

# 디렉토리 내 모든 파일에 대해 처리
for filename in os.listdir(directory):
    if filename.endswith(".mp3"):
        file_path = os.path.join(directory, filename)
        try:
            # 오디오 파일 로드 시도
            waveform, sr = torchaudio.load(file_path, normalize=True, format="mp3")
        except Exception as e:
            # 오류 발생 시 파일 이름을 저장
            print(f"Error loading file: {file_path} - {e}")
            error_files.append(filename)

# 결과 출력
print("Files with errors:")
for error_file in error_files:
    print(error_file)

# 결과를 파일로 저장 (선택 사항)
with open("error_files.txt", "w") as f:
    for error_file in error_files:
        f.write(error_file + "\n")
```

```python
import pandas as pd

# 오류가 발생한 파일 이름 리스트
error_file_path = "./error_files/train_error_files.txt"

# 파일에서 라인별로 읽어와 리스트로 변환
with open(error_file_path, "r") as f:
    error_files = [line.strip() for line in f]

# TSV 파일 경로
tsv_file = "./dataset/train/covost_v2.en_korean.train.tsv"

# TSV 파일 읽기
df = pd.read_csv(tsv_file, sep="\t")

# 오류가 발생한 파일 이름과 일치하는 행 제거
df_filtered = df[~df['path'].isin(error_files)]

# 결과 저장
output_file = "./dataset/train/covost_v2.en_korean.train.filtered.tsv"
df_filtered.to_csv(output_file, sep="\t", index=False)

print(f"Filtered TSV saved to {output_file}")
```
